```python
train_df['feature_10_ordinal'] = train_df['feature_10'].map({'low': 0, 'med': 1, 'high': 2})
train_df['feature_11_ordinal'] = train_df['feature_11'].map({'low': 0, 'med': 1, 'high': 2})
train_df['feature_12_ordinal'] = train_df['feature_12'].map({'low': 0, 'med': 1, 'high': 2})
test_df['feature_10_ordinal'] = test_df['feature_10'].map({'low': 0, 'med': 1, 'high': 2})
test_df['feature_11_ordinal'] = test_df['feature_11'].map({'low': 0, 'med': 1, 'high': 2})
test_df['feature_12_ordinal'] = test_df['feature_12'].map({'low': 0, 'med': 1, 'high': 2})
```

```python
train_df['target_class_feature_10_11_12'] = train_df.groupby(['feature_10', 'feature_11', 'feature_12'])['target'].transform(lambda x: x.mode().values[0])
train_df['target_class_feature_10_11'] = train_df.groupby(['feature_10', 'feature_11'])['target'].transform(lambda x: x.mode().values[0])
train_df['target_class_feature_10_12'] = train_df.groupby(['feature_10', 'feature_12'])['target'].transform(lambda x: x.mode().values[0])
train_df['target_class_feature_11_12'] = train_df.groupby(['feature_11', 'feature_12'])['target'].transform(lambda x: x.mode().values[0])
train_df['target_class_feature_10'] = train_df.groupby(['feature_10'])['target'].transform(lambda x: x.mode().values[0])
train_df['target_class_feature_11'] = train_df.groupby(['feature_11'])['target'].transform(lambda x: x.mode().values[0])
train_df['target_class_feature_12'] = train_df.groupby(['feature_12'])['target'].transform(lambda x: x.mode().values[0])
```

```python
for i, group in test_df.groupby(['feature_10', 'feature_11', 'feature_12']):
    test_df.loc[group.index, 'target_class_feature_10_11_12'] = train_df[(train_df['feature_10'] == group['feature_10'].values[0]) & (train_df['feature_11'] == group['feature_11'].values[0]) & (train_df['feature_12'] == group['feature_12'].values[0])]['target_class_feature_10_11_12'].values[0]
    test_df.loc[group.index, 'target_class_feature_10_11'] = train_df[(train_df['feature_10'] == group['feature_10'].values[0]) & (train_df['feature_11'] == group['feature_11'].values[0])]['target_class_feature_10_11'].values[0]
    test_df.loc[group.index, 'target_class_feature_10_12'] = train_df[(train_df['feature_10'] == group['feature_10'].values[0]) & (train_df['feature_12'] == group['feature_12'].values[0])]['target_class_feature_10_12'].values[0]
    test_df.loc[group.index, 'target_class_feature_11_12'] = train_df[(train_df['feature_11'] == group['feature_11'].values[0]) & (train_df['feature_12'] == group['feature_12'].values[0])]['target_class_feature_11_12'].values[0]
    test_df.loc[group.index, 'target_class_feature_10'] = train_df[(train_df['feature_10'] == group['feature_10'].values[0])]['target_class_feature_10'].values[0]
    test_df.loc[group.index, 'target_class_feature_11'] = train_df[(train_df['feature_11'] == group['feature_11'].values[0])]['target_class_feature_11'].values[0]
    test_df.loc[group.index, 'target_class_feature_12'] = train_df[(train_df['feature_12'] == group['feature_12'].values[0])]['target_class_feature_12'].values[0]
```

```python
for i, group in test_df.groupby(['feature_10', 'feature_11', 'feature_12']):
    if i in train_df.groupby(['feature_10', 'feature_11', 'feature_12']).groups:
        test_df.loc[group.index, 'target_class_feature_10_11_12'] = train_df.loc[train_df.groupby(['feature_10', 'feature_11', 'feature_12']).groups[i], 'target_class_feature_10_11_12'].values[0]
    elif i in train_df.groupby(['feature_10', 'feature_11']).groups:
        test_df.loc[group.index, 'target_class_feature_10_11'] = train_df.loc[train_df.groupby(['feature_10', 'feature_11']).groups[i], 'target_class_feature_10_11'].values[0]
    elif i in train_df.groupby(['feature_10', 'feature_12']).groups:
        test_df.loc[group.index, 'target_class_feature_10_12'] = train_df.loc[train_df.groupby(['feature_10', 'feature_12']).groups[i], 'target_class_feature_10_12'].values[0]
    elif i in train_df.groupby(['feature_11', 'feature_12']).groups:
        test_df.loc[group.index, 'target_class_feature_11_12'] = train_df.loc[train_df.groupby(['feature_11', 'feature_12']).groups[i], 'target_class_feature_11_12'].values[0]
    elif i in train_df.groupby(['feature_10']).groups:
        test_df.loc[group.index, 'target_class_feature_10'] = train_df.loc[train_df.groupby(['feature_10']).groups[i], 'target_class_feature_10'].values[0]
    elif i in train_df.groupby(['feature_11']).groups:
        test_df.loc[group.index, 'target_class_feature_11'] = train_df.loc[train_df.groupby(['feature_11']).groups[i], 'target_class_feature_11'].values[0]
    elif i in train_df.groupby(['feature_12']).groups:
        test_df.loc[group.index, 'target_class_feature_12'] = train_df.loc[train_df.groupby(['feature_12']).groups[i], 'target_class_feature_12'].values[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(train_df.drop(columns=['target', 'feature_10', 'feature_11', 'feature_12']), train_df['target'], test_size=0.2, random_state=42, shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, confusion_matrix
```

```python
rf = RandomForestClassifier(
    n_estimators=200,          # Number of trees in the forest
    criterion="gini",          # "gini" or "entropy" (for information gain)
    max_depth=10,            # Maximum depth of the tree (None means nodes are expanded until pure)
    min_samples_split=10,       # Minimum samples required to split a node
    min_samples_leaf=10,        # Minimum samples required at a leaf node
    min_weight_fraction_leaf=0.0, # Minimum fraction of sum of weights at a leaf node
    max_features="sqrt",       # Number of features to consider for split ('auto', 'sqrt', 'log2' or integer)
    max_leaf_nodes=None,
    bootstrap=True,          # Maximum number of leaf nodes
    min_impurity_decrease=0.0, # Minimum impurity decrease for node split
    oob_score=False,           # Out-of-bag estimates (only when bootstrap=True)
    n_jobs=-1,                 # Use all CPU cores
    random_state=42,           # Ensures reproducibility
    verbose=10,                 # 0 = silent, 1 = detailed output
    warm_start=False,          # Whether to reuse previous training data
    class_weight="balanced",         # Handling imbalanced classes ('balanced' or dictionary of class weights)
    ccp_alpha=0.0,             # Complexity parameter for pruning
)
```

```python
n_estimators_range = range(10, 500, 10)  # Number of trees to test
patience = 10  # Number of iterations to wait for improvement
best_accuracy = 0
best_n_estimators = 10
counter = 0

# Track validation accuracy for different numbers of trees
val_accuracies = []

for n_estimators in n_estimators_range:
    # Train model with 'n_estimators' trees
    rf = RandomForestClassifier(n_estimators=n_estimators, random_state=42, class_weight="balanced",
                                criterion="gini",          # "gini" or "entropy" (for information gain)
                                max_depth=15,            # Maximum depth of the tree (None means nodes are expanded until pure)
                                min_samples_split=10,       # Minimum samples required to split a node
                                min_samples_leaf=10,        # Minimum samples required at a leaf node
                                min_weight_fraction_leaf=0.0, # Minimum fraction of sum of weights at a leaf node
                                max_features='sqrt',       # Number of features to consider for split ('auto', 'sqrt', 'log2' or integer)
                                max_leaf_nodes=None,
                                bootstrap=True, 
                                min_impurity_decrease=0, # Minimum impurity decrease for node split
                                oob_score=False,           # Out-of-bag estimates (only when bootstrap=True)
                                n_jobs=-1,                 # Use all CPU cores
                                verbose=0,                 # 0 = silent, 1 = detailed output
                                warm_start=False,          # Whether to reuse previous training data
                                ccp_alpha=0)
    rf.fit(X_train, y_train)
    
    # Validate on train set
    y_train_pred = rf.predict(X_train)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    print(f"n_estimators: {n_estimators}, Train Accuracy: {train_accuracy:.4f}")
    
    # Validate on validation set
    y_val_pred = rf.predict(X_valid)
    val_accuracy = accuracy_score(y_valid, y_val_pred)
    val_accuracies.append(val_accuracy)

    print(f"n_estimators: {n_estimators}, Validation Accuracy: {val_accuracy:.4f}")

    # Check if validation accuracy improved
    if val_accuracy > best_accuracy:
        best_accuracy = val_accuracy
        best_n_estimators = n_estimators
        counter = 0  # Reset patience counter
    else:
        counter += 1
    
    # Stop training if no improvement for 'patience' iterations
    if counter >= patience:
        print(f"Early stopping at {n_estimators} trees (Best: {best_n_estimators} trees with {best_accuracy:.4f} accuracy)")
        break
```

```python
rf_full.fit(train_df.drop(columns=['target', 'feature_10', 'feature_11', 'feature_12']), train_df['target'])
```
