```python
def get_automatic_type(x):
    if "cvt" in x["transmission"].lower() or "variable" in x["transmission"].lower() or "cvt-f" in x["transmission"].lower():
        return "cvt"
    elif "electronically controlled" in x["transmission"].lower():
        return "electronically controlled"
    elif "overdrive" in x["transmission"].lower():
        return "overdrive"
    elif "dual shift" in x["transmission"].lower():
        return "dual shift"
    elif "DCT" in x["transmission"].lower():
        return "dct"
    elif "at/mt" in x["transmission"].lower():
        return "at/mt"
    elif "fixed" in x["transmission"].lower():
        return "fixed"
    elif "variable" in x["transmission"].lower():
        return "cvt"
    else:
        return "cvt"
```
